# Model Interpretability and Analysis

This notebook covers:
1. Feature importance for ALL ensemble models (LightGBM, CatBoost, XGBoost)
2. SHAP analysis comparison
3. Individual prediction explanations
4. Drift analysis between train and test
5. Error analysis

```python
import pickle
import numpy as np
import pandas as pd
import shap
from scipy.stats import ks_2samp
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# load data and all models
with open('../data/processed_data.pkl', 'rb') as f:
    data = pickle.load(f)

with open('../models/baseline_models.pkl', 'rb') as f:
    baseline = pickle.load(f)

X_train = data['X_train']
X_test = data['X_test']
y_train = data['y_train']
y_test = data['y_test']
feature_names = data['feature_names']

# load all models
lgb_model = baseline['models']['lightgbm']
cat_model = baseline['models']['catboost']
xgb_model = baseline['models']['xgboost']

print(f"Features: {len(feature_names)}")
print(f"Test samples: {len(y_test):,}")
print(f"Models loaded: LightGBM, CatBoost, XGBoost")
```

## 1. Feature Importance for All Ensemble Models

```python
# sample for SHAP (too slow on full data)
sample_size = 2000
np.random.seed(42)
sample_idx = np.random.choice(len(X_test), sample_size, replace=False)
X_sample = X_test[sample_idx]

print(f"Sample size for SHAP: {sample_size}")
```

```python
def get_feature_importance(model, model_name, X_sample, feature_names):
    """Get feature importance using SHAP or native importance."""
    try:
        # Try SHAP TreeExplainer
        explainer = shap.TreeExplainer(model)
        shap_values = explainer.shap_values(X_sample)
        
        # Handle different output formats
        if isinstance(shap_values, list):
            shap_values = shap_values[1]  # For binary classification, take positive class
        
        mean_abs_shap = np.abs(shap_values).mean(axis=0)
        importance_df = pd.DataFrame({
            'feature': feature_names,
            'importance': mean_abs_shap
        }).sort_values('importance', ascending=False)
        
        print(f"{model_name}: SHAP values computed successfully")
        return importance_df, shap_values, explainer
        
    except Exception as e:
        print(f"{model_name}: SHAP failed ({str(e)[:50]}...), using native importance")
        
        # Fallback to native feature importance
        importance = model.feature_importances_
        importance_df = pd.DataFrame({
            'feature': feature_names,
            'importance': importance
        }).sort_values('importance', ascending=False)
        
        return importance_df, None, None
```

```python
# Get feature importance for all models
print("Computing feature importance for all models...\n")

lgb_importance, lgb_shap, lgb_explainer = get_feature_importance(
    lgb_model, "LightGBM", X_sample, feature_names
)

cat_importance, cat_shap, cat_explainer = get_feature_importance(
    cat_model, "CatBoost", X_sample, feature_names
)

xgb_importance, xgb_shap, xgb_explainer = get_feature_importance(
    xgb_model, "XGBoost", X_sample, feature_names
)
```

### 1.1 Individual Model Feature Importance

```python
# Display top features for each model
print("="*70)
print("TOP 10 FEATURES BY MODEL")
print("="*70)

for name, imp_df in [('LightGBM', lgb_importance), ('CatBoost', cat_importance), ('XGBoost', xgb_importance)]:
    print(f"\n{name}:")
    print(imp_df.head(10).to_string(index=False))
```

```python
# Individual bar plots for each model
fig, axes = plt.subplots(1, 3, figsize=(18, 8))

for ax, (name, imp_df) in zip(axes, [('LightGBM', lgb_importance), 
                                       ('CatBoost', cat_importance), 
                                       ('XGBoost', xgb_importance)]):
    top_15 = imp_df.head(15)
    colors = plt.cm.Blues(np.linspace(0.4, 0.9, 15))[::-1]
    
    ax.barh(range(len(top_15)), top_15['importance'].values, color=colors)
    ax.set_yticks(range(len(top_15)))
    ax.set_yticklabels(top_15['feature'].values)
    ax.invert_yaxis()
    ax.set_xlabel('Feature Importance')
    ax.set_title(f'{name} - Top 15 Features')

plt.tight_layout()
plt.savefig('./plots/feature_importance_all_models.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 1.2 Combined Feature Importance Comparison

```python
# Normalize importance scores for comparison
lgb_norm = lgb_importance.copy()
lgb_norm['importance'] = lgb_norm['importance'] / lgb_norm['importance'].sum()
lgb_norm = lgb_norm.rename(columns={'importance': 'LightGBM'})

cat_norm = cat_importance.copy()
cat_norm['importance'] = cat_norm['importance'] / cat_norm['importance'].sum()
cat_norm = cat_norm.rename(columns={'importance': 'CatBoost'})

xgb_norm = xgb_importance.copy()
xgb_norm['importance'] = xgb_norm['importance'] / xgb_norm['importance'].sum()
xgb_norm = xgb_norm.rename(columns={'importance': 'XGBoost'})

# Merge all
combined = lgb_norm.merge(cat_norm, on='feature').merge(xgb_norm, on='feature')
combined['avg_importance'] = combined[['LightGBM', 'CatBoost', 'XGBoost']].mean(axis=1)
combined = combined.sort_values('avg_importance', ascending=False)

print("Top 15 Features (Average across all models):")
print(combined.head(15)[['feature', 'LightGBM', 'CatBoost', 'XGBoost', 'avg_importance']].round(4).to_string(index=False))
```

```python
# Grouped bar chart comparison
fig, ax = plt.subplots(figsize=(14, 10))

top_15_features = combined.head(15)['feature'].values
x = np.arange(len(top_15_features))
width = 0.25

# Get values for top 15 features
lgb_vals = combined.head(15)['LightGBM'].values
cat_vals = combined.head(15)['CatBoost'].values
xgb_vals = combined.head(15)['XGBoost'].values

bars1 = ax.barh(x - width, lgb_vals, width, label='LightGBM', color='#2ecc71')
bars2 = ax.barh(x, cat_vals, width, label='CatBoost', color='#3498db')
bars3 = ax.barh(x + width, xgb_vals, width, label='XGBoost', color='#e74c3c')

ax.set_yticks(x)
ax.set_yticklabels(top_15_features)
ax.invert_yaxis()
ax.set_xlabel('Normalized Feature Importance')
ax.set_title('Feature Importance Comparison Across All Ensemble Models')
ax.legend(loc='lower right')

plt.tight_layout()
plt.savefig('./plots/feature_importance_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 1.3 Feature Importance Correlation Between Models

```python
# Check correlation between model importance rankings
from scipy.stats import spearmanr

# Add rank columns
combined['rank_lgb'] = combined['LightGBM'].rank(ascending=False)
combined['rank_cat'] = combined['CatBoost'].rank(ascending=False)
combined['rank_xgb'] = combined['XGBoost'].rank(ascending=False)

# Spearman correlations
corr_lgb_cat, _ = spearmanr(combined['rank_lgb'], combined['rank_cat'])
corr_lgb_xgb, _ = spearmanr(combined['rank_lgb'], combined['rank_xgb'])
corr_cat_xgb, _ = spearmanr(combined['rank_cat'], combined['rank_xgb'])

print("Feature Importance Rank Correlation (Spearman):")
print(f"  LightGBM vs CatBoost: {corr_lgb_cat:.4f}")
print(f"  LightGBM vs XGBoost:  {corr_lgb_xgb:.4f}")
print(f"  CatBoost vs XGBoost:  {corr_cat_xgb:.4f}")
print(f"\nInterpretation: High correlation means models agree on important features")
```

### 1.4 SHAP Summary Plots (if available)

```python
# SHAP summary plots for models where SHAP succeeded - Each in separate figure for clarity
shap_available = []

if lgb_shap is not None:
    shap_available.append(('LightGBM', lgb_shap))
if cat_shap is not None:
    shap_available.append(('CatBoost', cat_shap))
if xgb_shap is not None:
    shap_available.append(('XGBoost', xgb_shap))

if len(shap_available) > 0:
    print(f"Generating SHAP Summary Plots for {len(shap_available)} models...")
    print("=" * 60)
    
    # Create separate figures for each model (less congested)
    for name, shap_vals in shap_available:
        fig, ax = plt.subplots(figsize=(12, 10))
        plt.sca(ax)
        shap.summary_plot(shap_vals, X_sample, feature_names=feature_names, 
                         show=False, max_display=20)
        plt.title(f'{name} SHAP Summary - Top 20 Features', fontsize=14, pad=20)
        plt.tight_layout()
        plt.savefig(f'./plots/shap_summary_{name.lower()}.png', dpi=150, bbox_inches='tight')
        plt.show()
        print(f"\n{name} SHAP plot saved to ./plots/shap_summary_{name.lower()}.png")
    
    # Also create a combined view with more space
    n_models = len(shap_available)
    fig, axes = plt.subplots(n_models, 1, figsize=(14, 8*n_models))
    if n_models == 1:
        axes = [axes]
    
    for ax, (name, shap_vals) in zip(axes, shap_available):
        plt.sca(ax)
        shap.summary_plot(shap_vals, X_sample, feature_names=feature_names, 
                         show=False, max_display=15)
        ax.set_title(f'{name} SHAP Summary', fontsize=12)
    
    plt.tight_layout()
    plt.savefig('./plots/shap_summary_all_models.png', dpi=150, bbox_inches='tight')
    plt.show()
else:
    print("SHAP summary plots not available (all models fell back to native importance)")
    print("See bar plots above for feature importance visualization")
```

## 2. Key Feature Analysis

```python
# Identify consistently important features across all models
top10_lgb = set(lgb_importance.head(10)['feature'])
top10_cat = set(cat_importance.head(10)['feature'])
top10_xgb = set(xgb_importance.head(10)['feature'])

# Features in top 10 of all models
common_top10 = top10_lgb & top10_cat & top10_xgb

print("Features in Top 10 of ALL models:")
for feat in common_top10:
    lgb_rank = lgb_importance[lgb_importance['feature'] == feat].index[0] + 1
    cat_rank = cat_importance[cat_importance['feature'] == feat].index[0] + 1
    xgb_rank = xgb_importance[xgb_importance['feature'] == feat].index[0] + 1
    print(f"  {feat}: LGB #{lgb_rank}, Cat #{cat_rank}, XGB #{xgb_rank}")

print(f"\nTotal common features: {len(common_top10)}")
```

```python
# Feature category breakdown
def categorize_feature(feat):
    if any(x in feat for x in ['amount', 'ratio', 'sum', 'mean', 'std', 'log']):
        return 'Amount-based'
    elif any(x in feat for x in ['tx_count', 'velocity']):
        return 'Velocity'
    elif any(x in feat for x in ['hour', 'day', 'month', 'time', 'gap']):
        return 'Temporal'
    elif any(x in feat for x in ['channel', 'location', 'category', 'encoded']):
        return 'Categorical'
    else:
        return 'Other'

combined['category'] = combined['feature'].apply(categorize_feature)

# Category importance by model
cat_summary = combined.groupby('category')[['LightGBM', 'CatBoost', 'XGBoost']].sum()
print("\nFeature Category Importance (sum of normalized importance):")
print(cat_summary.round(4))
```

```python
# Pie chart for feature categories
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

colors = ['#3498db', '#2ecc71', '#e74c3c', '#f39c12', '#9b59b6']

for ax, model in zip(axes, ['LightGBM', 'CatBoost', 'XGBoost']):
    ax.pie(cat_summary[model], labels=cat_summary.index, autopct='%1.1f%%', colors=colors)
    ax.set_title(f'{model} - Feature Category Distribution')

plt.tight_layout()
plt.savefig('./plots/feature_category_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. Drift Analysis

```python
# KS test for feature drift
drift_results = []

for i, feat in enumerate(feature_names):
    stat, pval = ks_2samp(X_train[:, i], X_test[:, i])
    drift_results.append({
        'feature': feat,
        'ks_statistic': stat,
        'p_value': pval,
        'significant': pval < 0.05
    })

drift_df = pd.DataFrame(drift_results).sort_values('ks_statistic', ascending=False)

print("Top 10 Features with Highest Drift:")
print(drift_df.head(10).to_string(index=False))

n_drifted = drift_df['significant'].sum()
print(f"\nFeatures with significant drift: {n_drifted}/{len(feature_names)}")
```

```python
# fraud rate drift
train_fraud_rate = y_train.mean()
test_fraud_rate = y_test.mean()

print(f"\nFraud Rate:")
print(f"  Train: {train_fraud_rate:.4%}")
print(f"  Test:  {test_fraud_rate:.4%}")
print(f"  Ratio: {test_fraud_rate/train_fraud_rate:.2f}x")
```

## 4. Error Analysis

```python
# use optimal threshold for best model (XGBoost)
import os

y_proba = baseline['predictions']['xgboost']

if os.path.exists('../models/ensemble_results.pkl'):
    with open('../models/ensemble_results.pkl', 'rb') as f:
        ens_results = pickle.load(f)
    threshold = ens_results['thresholds']['xgboost']
else:
    from sklearn.metrics import precision_recall_curve, f1_score as f1
    precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
    f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)
    threshold = thresholds[np.argmax(f1_scores[:-1])]
    print(f"Calculated optimal threshold: {threshold:.4f}")

y_pred = (y_proba >= threshold).astype(int)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

print(f"Confusion Matrix (threshold={threshold:.4f}):")
print(f"  True Positives:  {tp:,}")
print(f"  False Positives: {fp:,}")
print(f"  True Negatives:  {tn:,}")
print(f"  False Negatives: {fn:,}")
```

```python
# analyze false negatives (missed fraud)
fn_mask = (y_test == 1) & (y_pred == 0)
fn_proba = y_proba[fn_mask]

print(f"\nFalse Negatives Analysis:")
print(f"  Count: {fn_mask.sum():,}")
print(f"  Mean probability: {fn_proba.mean():.4f}")
print(f"  Max probability: {fn_proba.max():.4f}")
```

```python
# probability distribution by outcome
fig, ax = plt.subplots(figsize=(10, 5))

ax.hist(y_proba[y_test == 0], bins=50, alpha=0.6, label='Legitimate', density=True)
ax.hist(y_proba[y_test == 1], bins=50, alpha=0.6, label='Fraud', density=True)
ax.axvline(threshold, color='r', linestyle='--', label=f'Threshold ({threshold:.3f})')

ax.set_xlabel('Predicted Probability')
ax.set_ylabel('Density')
ax.set_title('Score Distribution by Class')
ax.legend()

plt.tight_layout()
plt.savefig('./plots/score_distribution.png', dpi=150)
plt.show()
```

## 5. Save Analysis

```python
# save analysis
analysis_data = {
    'feature_importance': {
        'lightgbm': lgb_importance,
        'catboost': cat_importance,
        'xgboost': xgb_importance,
        'combined': combined
    },
    'drift_analysis': drift_df,
    'shap_values': {
        'lightgbm': lgb_shap,
        'catboost': cat_shap,
        'xgboost': xgb_shap
    },
    'sample_idx': sample_idx
}

with open('../models/interpretability_analysis.pkl', 'wb') as f:
    pickle.dump(analysis_data, f)

print("Analysis saved to ../models/interpretability_analysis.pkl")
```

2. **Feature Categories:**

   - Velocity: Second most important (~20-30%)
   - Temporal: Moderate importance (~10-15%)
   - Categorical: Lower but still useful

3. **Drift:**
   - Some features show statistical drift between train and test
   - Fraud rate stable (train: 0.30%, test: 0.29%)

4. **Errors:**
   - False negatives have low predicted probability
   - These are genuinely hard-to-detect fraud cases
