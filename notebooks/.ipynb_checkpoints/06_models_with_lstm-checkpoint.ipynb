# Models with LSTM Features

Train LightGBM, CatBoost, XGBoost on LSTM-enhanced features (61 original + 64 LSTM = 125 total).

**Target Metrics:**
- F1 >= 0.90
- AUC-ROC >= 0.95
- FPR < 0.1%

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import lightgbm as lgb
from catboost import CatBoostClassifier
import xgboost as xgb
from sklearn.metrics import (roc_auc_score, f1_score, precision_score, recall_score,
                             confusion_matrix, precision_recall_curve, average_precision_score,
                             roc_curve)
import warnings
warnings.filterwarnings('ignore')
```

## 1. Load LSTM-Enhanced Data

```python
with open('../data/processed_data_lstm.pkl', 'rb') as f:
    data = pickle.load(f)

X_train = data['X_train']
X_test = data['X_test']
y_train = data['y_train']
y_test = data['y_test']
feature_names = data['feature_names']
scale_pos_weight = data['scale_pos_weight']

# handle NaN/Inf
X_train = np.nan_to_num(X_train, nan=0.0, posinf=0.0, neginf=0.0)
X_test = np.nan_to_num(X_test, nan=0.0, posinf=0.0, neginf=0.0)

print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
print(f"Train fraud rate: {y_train.mean()*100:.3f}%")
print(f"Test fraud rate: {y_test.mean()*100:.3f}%")
print(f"Features: {len(feature_names)} ({len(data['lstm_embedding_cols'])} LSTM)")
print(f"Scale pos weight: {scale_pos_weight:.1f}")
```

## 2. Helper Functions

```python
def find_optimal_threshold(y_true, y_proba):
    """Find threshold maximizing F1."""
    precision, recall, thresholds = precision_recall_curve(y_true, y_proba)
    f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)
    best_idx = np.argmax(f1_scores[:-1])
    return thresholds[best_idx]

def evaluate(y_true, y_proba, threshold=None):
    """Evaluate model."""
    if threshold is None:
        threshold = find_optimal_threshold(y_true, y_proba)

    y_pred = (y_proba >= threshold).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

    return {
        'auc': roc_auc_score(y_true, y_proba),
        'pr_auc': average_precision_score(y_true, y_proba),
        'f1': f1_score(y_true, y_pred),
        'precision': precision_score(y_true, y_pred),
        'recall': recall_score(y_true, y_pred),
        'fpr': fp / (fp + tn),
        'threshold': threshold,
        'tp': tp, 'fp': fp, 'tn': tn, 'fn': fn
    }
```

## 3. Train LightGBM

```python
lgb_params = {
    'objective': 'binary',
    'metric': ['auc', 'binary_logloss'],
    'boosting_type': 'gbdt',
    'num_leaves': 128,
    'learning_rate': 0.03,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'scale_pos_weight': scale_pos_weight,
    'verbose': -1,
    'n_jobs': -1,
    'random_state': 42
}

lgb_train = lgb.Dataset(X_train, y_train, feature_name=feature_names)
lgb_valid = lgb.Dataset(X_test, y_test, reference=lgb_train)

print("Training LightGBM...")
lgb_model = lgb.train(
    lgb_params,
    lgb_train,
    num_boost_round=1000,
    valid_sets=[lgb_train, lgb_valid],
    valid_names=['train', 'valid'],
    callbacks=[lgb.early_stopping(50), lgb.log_evaluation(100)]
)

lgb_proba = lgb_model.predict(X_test)
lgb_results = evaluate(y_test, lgb_proba)

print(f"\nLightGBM: AUC={lgb_results['auc']:.4f}, F1={lgb_results['f1']:.4f}")
```

## 4. Train CatBoost

```python
cat_model = CatBoostClassifier(
    iterations=1000,
    learning_rate=0.03,
    depth=8,
    l2_leaf_reg=3,
    class_weights=[1, scale_pos_weight],
    random_seed=42,
    verbose=100,
    early_stopping_rounds=50
)

print("Training CatBoost...")
cat_model.fit(X_train, y_train, eval_set=[(X_test, y_test)])

cat_proba = cat_model.predict_proba(X_test)[:, 1]
cat_results = evaluate(y_test, cat_proba)

print(f"\nCatBoost: AUC={cat_results['auc']:.4f}, F1={cat_results['f1']:.4f}")
```

## 5. Train XGBoost

```python
xgb_model = xgb.XGBClassifier(
    n_estimators=1000,
    max_depth=8,
    learning_rate=0.03,
    scale_pos_weight=scale_pos_weight,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.1,
    reg_lambda=1.0,
    random_state=42,
    n_jobs=-1,
    eval_metric='auc',
    early_stopping_rounds=50
)

print("Training XGBoost...")
xgb_model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=100)

xgb_proba = xgb_model.predict_proba(X_test)[:, 1]
xgb_results = evaluate(y_test, xgb_proba)

print(f"\nXGBoost: AUC={xgb_results['auc']:.4f}, F1={xgb_results['f1']:.4f}")
```

## 6. Model Comparison

```python
results = {
    'LightGBM': lgb_results,
    'CatBoost': cat_results,
    'XGBoost': xgb_results
}

print(f"{'Model':<12} {'AUC':>8} {'PR-AUC':>8} {'F1':>8} {'Precision':>10} {'Recall':>8} {'FPR':>8}")
print("-" * 70)
for name, res in results.items():
    print(f"{name:<12} {res['auc']:>8.4f} {res['pr_auc']:>8.4f} {res['f1']:>8.4f} "
          f"{res['precision']:>10.4f} {res['recall']:>8.4f} {res['fpr']:>8.4%}")

best_name = max(results, key=lambda x: results[x]['f1'])
best = results[best_name]
print(f"\nBest model: {best_name}")
```

```python
# target metrics check
print("\nTARGET METRICS CHECK")
print("=" * 40)
print(f"  F1 >= 0.90:  {best['f1']:.4f} {'PASS' if best['f1'] >= 0.90 else 'FAIL'}")
print(f"  AUC >= 0.95: {best['auc']:.4f} {'PASS' if best['auc'] >= 0.95 else 'FAIL'}")
print(f"  FPR < 0.1%:  {best['fpr']:.4%} {'PASS' if best['fpr'] < 0.001 else 'FAIL'}")
```

```python
# ROC curves
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

colors = ['#2E86AB', '#A23B72', '#F18F01']
models = [('LightGBM', lgb_proba), ('CatBoost', cat_proba), ('XGBoost', xgb_proba)]

for (name, proba), color in zip(models, colors):
    fpr, tpr, _ = roc_curve(y_test, proba)
    auc = results[name]['auc']
    axes[0].plot(fpr, tpr, label=f'{name} (AUC={auc:.4f})', color=color)

axes[0].plot([0, 1], [0, 1], 'k--', alpha=0.3)
axes[0].set_xlabel('False Positive Rate')
axes[0].set_ylabel('True Positive Rate')
axes[0].set_title('ROC Curves')
axes[0].legend()

for (name, proba), color in zip(models, colors):
    precision, recall, _ = precision_recall_curve(y_test, proba)
    pr_auc = results[name]['pr_auc']
    axes[1].plot(recall, precision, label=f'{name} (PR-AUC={pr_auc:.4f})', color=color)

axes[1].set_xlabel('Recall')
axes[1].set_ylabel('Precision')
axes[1].set_title('Precision-Recall Curves')
axes[1].legend()

plt.tight_layout()
plt.savefig('plots/lstm_model_curves.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 7. Feature Importance

```python
import seaborn as sns

importance = cat_model.get_feature_importance()
feat_imp = sorted(zip(feature_names, importance), key=lambda x: x[1], reverse=True)

print("Top 20 Features:")
print("-" * 40)
for i, (name, imp) in enumerate(feat_imp[:20], 1):
    marker = " [LSTM]" if name.startswith('lstm_emb') else ""
    print(f"  {i:2d}. {name}: {imp:.2f}{marker}")
```

```python
# LSTM embedding contribution analysis
lstm_importance = sum(imp for name, imp in feat_imp if name.startswith('lstm_emb'))
original_importance = sum(imp for name, imp in feat_imp if not name.startswith('lstm_emb'))
total_importance = lstm_importance + original_importance

print(f"\nFeature Type Contribution:")
print(f"  Original Features: {original_importance:.2f} ({original_importance/total_importance*100:.1f}%)")
print(f"  LSTM Embeddings:   {lstm_importance:.2f} ({lstm_importance/total_importance*100:.1f}%)")

# Pie chart
fig, ax = plt.subplots(figsize=(8, 6))
sizes = [original_importance, lstm_importance]
labels = [f'Original Features\n({original_importance/total_importance*100:.1f}%)',
          f'LSTM Embeddings\n({lstm_importance/total_importance*100:.1f}%)']
colors = ['#3498db', '#e74c3c']
explode = (0, 0.05)

ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='',
       shadow=True, startangle=90)
ax.set_title('Feature Type Contribution to CatBoost Model')

plt.tight_layout()
plt.savefig('plots/lstm_contribution_pie.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Feature importance bar chart
fig, ax = plt.subplots(figsize=(12, 8))

top_20 = feat_imp[:20]
names = [f[0] for f in top_20]
values = [f[1] for f in top_20]
colors = ['#e74c3c' if n.startswith('lstm_emb') else '#3498db' for n in names]

bars = ax.barh(range(len(names)), values, color=colors)
ax.set_yticks(range(len(names)))
ax.set_yticklabels(names)
ax.invert_yaxis()
ax.set_xlabel('Feature Importance')
ax.set_title('Top 20 Features (LSTM-Enhanced Model)\nRed = LSTM Embedding, Blue = Original Feature')

# Add legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#3498db', label='Original Features'),
                   Patch(facecolor='#e74c3c', label='LSTM Embeddings')]
ax.legend(handles=legend_elements, loc='lower right')

plt.tight_layout()
plt.savefig('plots/lstm_feature_importance_catboost.png', dpi=150, bbox_inches='tight')
plt.show()
```
