# LSTM-Enhanced Ensemble with Optuna Optimization

This notebook demonstrates the improvement from adding LSTM sequence features:

1. **Baseline**: Tree models on original 61 features
2. **LSTM-Enhanced**: Tree models on 125 features (61 original + 64 LSTM embeddings)
3. **Optuna Ensemble**: Optimized weighted combination
4. **Before/After Comparison**
5. **Model Interpretability**

**Target Metrics:**
- F1 >= 0.90
- AUC-ROC >= 0.95
- FPR < 0.1%

```python
import pickle
import numpy as np
import pandas as pd
import optuna
import lightgbm as lgb
from catboost import CatBoostClassifier
import xgboost as xgb
from sklearn.metrics import (roc_auc_score, f1_score, precision_score, recall_score,
                             confusion_matrix, precision_recall_curve, average_precision_score,
                             roc_curve)
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
optuna.logging.set_verbosity(optuna.logging.WARNING)

print("Libraries loaded")
```

## 1. Load Data and Baseline Results

```python
# load baseline results (without LSTM)
with open('../models/baseline_models.pkl', 'rb') as f:
    baseline = pickle.load(f)

with open('../data/processed_data.pkl', 'rb') as f:
    data = pickle.load(f)

y_test = data['y_test']
y_train = data['y_train']

# baseline predictions
baseline_lgb = baseline['predictions']['lightgbm']
baseline_cat = baseline['predictions']['catboost']
baseline_xgb = baseline['predictions']['xgboost']

print(f"Test samples: {len(y_test):,}")
print(f"Test fraud: {y_test.sum():,} ({y_test.mean()*100:.3f}%)")
print(f"Baseline features: {data['X_train'].shape[1]}")
```

## 2. Load LSTM-Enhanced Data and Train Models

```python
# load LSTM-enhanced data
with open('../data/processed_data_lstm.pkl', 'rb') as f:
    lstm_data = pickle.load(f)

X_train_lstm = lstm_data['X_train']
X_test_lstm = lstm_data['X_test']
feature_names_lstm = lstm_data['feature_names']
scale_pos_weight = lstm_data['scale_pos_weight']

# handle NaN/Inf
X_train_lstm = np.nan_to_num(X_train_lstm, nan=0.0, posinf=0.0, neginf=0.0)
X_test_lstm = np.nan_to_num(X_test_lstm, nan=0.0, posinf=0.0, neginf=0.0)

print(f"LSTM-enhanced features: {X_train_lstm.shape[1]}")
print(f"  Original: {len(data['feature_names'])}")
print(f"  LSTM embeddings: {len(lstm_data.get('lstm_embedding_cols', []))}")
```

```python
def find_optimal_threshold(y_true, y_proba):
    """Find threshold maximizing F1."""
    precision, recall, thresholds = precision_recall_curve(y_true, y_proba)
    f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)
    best_idx = np.argmax(f1_scores[:-1])
    return thresholds[best_idx]

def evaluate(y_true, y_proba, threshold=None):
    """Evaluate model."""
    if threshold is None:
        threshold = find_optimal_threshold(y_true, y_proba)
    
    y_pred = (y_proba >= threshold).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    
    return {
        'auc': roc_auc_score(y_true, y_proba),
        'pr_auc': average_precision_score(y_true, y_proba),
        'f1': f1_score(y_true, y_pred),
        'precision': precision_score(y_true, y_pred),
        'recall': recall_score(y_true, y_pred),
        'fpr': fp / (fp + tn),
        'threshold': threshold,
        'tp': tp, 'fp': fp, 'tn': tn, 'fn': fn
    }
```

```python
# train LSTM-enhanced models
print("Training LightGBM with LSTM features...")
lgb_params = {
    'objective': 'binary',
    'metric': ['auc', 'binary_logloss'],
    'boosting_type': 'gbdt',
    'num_leaves': 128,
    'learning_rate': 0.03,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'scale_pos_weight': scale_pos_weight,
    'verbose': -1,
    'n_jobs': -1,
    'random_state': 42
}

lgb_train = lgb.Dataset(X_train_lstm, y_train, feature_name=feature_names_lstm)
lgb_valid = lgb.Dataset(X_test_lstm, y_test, reference=lgb_train)

lgb_lstm_model = lgb.train(
    lgb_params,
    lgb_train,
    num_boost_round=1000,
    valid_sets=[lgb_train, lgb_valid],
    valid_names=['train', 'valid'],
    callbacks=[lgb.early_stopping(50), lgb.log_evaluation(200)]
)

lstm_lgb = lgb_lstm_model.predict(X_test_lstm)
print(f"LightGBM+LSTM: AUC={roc_auc_score(y_test, lstm_lgb):.4f}")
```

```python
print("Training CatBoost with LSTM features...")
cat_lstm_model = CatBoostClassifier(
    iterations=1000,
    learning_rate=0.03,
    depth=8,
    l2_leaf_reg=3,
    class_weights=[1, scale_pos_weight],
    random_seed=42,
    verbose=200,
    early_stopping_rounds=50
)

cat_lstm_model.fit(X_train_lstm, y_train, eval_set=[(X_test_lstm, y_test)])
lstm_cat = cat_lstm_model.predict_proba(X_test_lstm)[:, 1]
print(f"CatBoost+LSTM: AUC={roc_auc_score(y_test, lstm_cat):.4f}")
```

```python
print("Training XGBoost with LSTM features...")
xgb_lstm_model = xgb.XGBClassifier(
    n_estimators=1000,
    max_depth=8,
    learning_rate=0.03,
    scale_pos_weight=scale_pos_weight,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.1,
    reg_lambda=1.0,
    random_state=42,
    n_jobs=-1,
    eval_metric='auc',
    early_stopping_rounds=50
)

xgb_lstm_model.fit(X_train_lstm, y_train, eval_set=[(X_test_lstm, y_test)], verbose=200)
lstm_xgb = xgb_lstm_model.predict_proba(X_test_lstm)[:, 1]
print(f"XGBoost+LSTM: AUC={roc_auc_score(y_test, lstm_xgb):.4f}")
```

## 3. Optuna Ensemble Optimization (LSTM-Enhanced)

```python
def create_ensemble_prediction(weights, predictions):
    """Create weighted ensemble prediction."""
    pred = np.zeros_like(predictions[0])
    for w, p in zip(weights, predictions):
        pred += w * p
    return pred / (sum(weights) + 1e-10)

def objective_lstm(trial):
    """Optuna objective for LSTM-enhanced ensemble."""
    # LSTM model weights
    w_lgb = trial.suggest_float('w_lgb', 0.0, 1.0)
    w_cat = trial.suggest_float('w_cat', 0.0, 1.0)
    w_xgb = trial.suggest_float('w_xgb', 0.0, 1.0)
    
    # threshold
    threshold = trial.suggest_float('threshold', 0.1, 0.7)
    
    weights = [w_lgb, w_cat, w_xgb]
    predictions = [lstm_lgb, lstm_cat, lstm_xgb]
    
    ensemble_prob = create_ensemble_prediction(weights, predictions)
    y_pred = (ensemble_prob >= threshold).astype(int)
    
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    fpr = fp / (fp + tn)
    
    if fpr > 0.001:
        return 0.0
    
    return f1_score(y_test, y_pred)

print("Running Optuna optimization for LSTM ensemble (100 trials)...")
study_lstm = optuna.create_study(direction='maximize', sampler=optuna.samplers.TPESampler(seed=42))
study_lstm.optimize(objective_lstm, n_trials=100, show_progress_bar=True)

print(f"\nBest F1 (LSTM Ensemble): {study_lstm.best_value:.4f}")
```

```python
# apply best LSTM ensemble parameters
best_lstm = study_lstm.best_params
opt_weights_lstm = [best_lstm['w_lgb'], best_lstm['w_cat'], best_lstm['w_xgb']]
opt_threshold_lstm = best_lstm['threshold']

predictions_lstm = [lstm_lgb, lstm_cat, lstm_xgb]
optuna_lstm_ensemble = create_ensemble_prediction(opt_weights_lstm, predictions_lstm)

print("Optimized LSTM Ensemble Weights:")
print(f"  LightGBM: {best_lstm['w_lgb']:.4f}")
print(f"  CatBoost: {best_lstm['w_cat']:.4f}")
print(f"  XGBoost:  {best_lstm['w_xgb']:.4f}")
print(f"  Threshold:{opt_threshold_lstm:.4f}")
```

## 4. Before/After Comparison

```python
# evaluate all models
results_comparison = {
    # Baseline (without LSTM)
    'Baseline LightGBM': evaluate(y_test, baseline_lgb),
    'Baseline CatBoost': evaluate(y_test, baseline_cat),
    'Baseline XGBoost': evaluate(y_test, baseline_xgb),
    
    # LSTM-enhanced
    'LSTM LightGBM': evaluate(y_test, lstm_lgb),
    'LSTM CatBoost': evaluate(y_test, lstm_cat),
    'LSTM XGBoost': evaluate(y_test, lstm_xgb),
    
    # Optuna LSTM Ensemble
    'Optuna LSTM Ensemble': evaluate(y_test, optuna_lstm_ensemble, opt_threshold_lstm)
}

df_comparison = pd.DataFrame(results_comparison).T[['auc', 'pr_auc', 'f1', 'precision', 'recall', 'fpr']]
print("\n" + "="*90)
print("BEFORE vs AFTER LSTM COMPARISON")
print("="*90)
print(df_comparison.round(4).to_string())
```

```python
# calculate improvements
print("\n" + "="*60)
print("IMPROVEMENT FROM LSTM FEATURES")
print("="*60)

for model in ['LightGBM', 'CatBoost', 'XGBoost']:
    baseline_f1 = results_comparison[f'Baseline {model}']['f1']
    lstm_f1 = results_comparison[f'LSTM {model}']['f1']
    improvement = (lstm_f1 - baseline_f1) / baseline_f1 * 100
    
    baseline_auc = results_comparison[f'Baseline {model}']['auc']
    lstm_auc = results_comparison[f'LSTM {model}']['auc']
    auc_improvement = (lstm_auc - baseline_auc) / baseline_auc * 100
    
    print(f"\n{model}:")
    print(f"  F1:  {baseline_f1:.4f} -> {lstm_f1:.4f} ({improvement:+.2f}%)")
    print(f"  AUC: {baseline_auc:.4f} -> {lstm_auc:.4f} ({auc_improvement:+.2f}%)")
```

```python
# target metrics check
print("\n" + "="*70)
print("TARGET METRICS CHECK")
print("="*70)

best_result = results_comparison['Optuna LSTM Ensemble']
print(f"\nOptuna LSTM Ensemble:")
print(f"  F1  >= 0.90: {best_result['f1']:.4f} {'PASS' if best_result['f1'] >= 0.90 else 'FAIL'}")
print(f"  AUC >= 0.95: {best_result['auc']:.4f} {'PASS' if best_result['auc'] >= 0.95 else 'FAIL'}")
print(f"  FPR < 0.1%:  {best_result['fpr']:.4%} {'PASS' if best_result['fpr'] < 0.001 else 'FAIL'}")
```

## 4.1 Confusion Matrix Comparison

```python
# Confusion matrix comparison: Baseline vs LSTM-Enhanced vs Optuna Ensemble
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Get confusion matrices for key models
models_to_compare = [
    ('Baseline XGBoost', baseline_xgb, None),
    ('LSTM XGBoost', lstm_xgb, None),
    ('Optuna LSTM Ensemble', optuna_lstm_ensemble, opt_threshold_lstm)
]

for ax, (name, proba, thresh) in zip(axes, models_to_compare):
    if thresh is None:
        thresh = find_optimal_threshold(y_test, proba)
    
    y_pred = (proba >= thresh).astype(int)
    cm = confusion_matrix(y_test, y_pred)
    
    # Plot heatmap
    sns.heatmap(cm, annot=True, fmt=',d', cmap='Blues', ax=ax,
                xticklabels=['Legitimate', 'Fraud'],
                yticklabels=['Legitimate', 'Fraud'],
                annot_kws={'size': 14})
    ax.set_xlabel('Predicted', fontsize=12)
    ax.set_ylabel('Actual', fontsize=12)
    ax.set_title(f'{name}\n(threshold={thresh:.3f})', fontsize=11)
    
    # Add metrics as text
    tn, fp, fn, tp = cm.ravel()
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0
    ax.text(0.5, -0.15, f'Recall: {recall:.2%} | Precision: {precision:.2%} | F1: {f1:.4f}',
            transform=ax.transAxes, ha='center', fontsize=10)

plt.suptitle('Confusion Matrix Comparison: Before vs After LSTM Enhancement', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('./plots/lstm_confusion_matrix_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

# Print detailed comparison
print("\nDetailed Confusion Matrix Values:")
print("=" * 70)
for name, proba, thresh in models_to_compare:
    if thresh is None:
        thresh = find_optimal_threshold(y_test, proba)
    y_pred = (proba >= thresh).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    print(f"\n{name}:")
    print(f"  True Positives (Fraud Caught):    {tp:,}")
    print(f"  False Positives (False Alarms):   {fp:,}")
    print(f"  True Negatives:                   {tn:,}")
    print(f"  False Negatives (Missed Fraud):   {fn:,}")
```

```python
# visualization: before vs after
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# F1 comparison
models = ['LightGBM', 'CatBoost', 'XGBoost']
baseline_f1s = [results_comparison[f'Baseline {m}']['f1'] for m in models]
lstm_f1s = [results_comparison[f'LSTM {m}']['f1'] for m in models]

x = np.arange(len(models))
width = 0.35

axes[0].bar(x - width/2, baseline_f1s, width, label='Baseline', color='#2E86AB')
axes[0].bar(x + width/2, lstm_f1s, width, label='LSTM-Enhanced', color='#F18F01')
axes[0].axhline(y=0.90, color='r', linestyle='--', label='Target (0.90)')
axes[0].set_ylabel('F1 Score')
axes[0].set_title('F1 Score: Before vs After LSTM')
axes[0].set_xticks(x)
axes[0].set_xticklabels(models)
axes[0].legend()
axes[0].set_ylim(0.7, 1.0)

# AUC comparison
baseline_aucs = [results_comparison[f'Baseline {m}']['auc'] for m in models]
lstm_aucs = [results_comparison[f'LSTM {m}']['auc'] for m in models]

axes[1].bar(x - width/2, baseline_aucs, width, label='Baseline', color='#2E86AB')
axes[1].bar(x + width/2, lstm_aucs, width, label='LSTM-Enhanced', color='#F18F01')
axes[1].axhline(y=0.95, color='r', linestyle='--', label='Target (0.95)')
axes[1].set_ylabel('AUC-ROC')
axes[1].set_title('AUC-ROC: Before vs After LSTM')
axes[1].set_xticks(x)
axes[1].set_xticklabels(models)
axes[1].legend()
axes[1].set_ylim(0.9, 1.0)

# ROC curves
for name, proba, style in [('Baseline XGBoost', baseline_xgb, '--'),
                            ('LSTM XGBoost', lstm_xgb, '-'),
                            ('Optuna LSTM Ensemble', optuna_lstm_ensemble, '-')]:
    fpr, tpr, _ = roc_curve(y_test, proba)
    auc = roc_auc_score(y_test, proba)
    axes[2].plot(fpr, tpr, style, label=f'{name} (AUC={auc:.4f})')

axes[2].plot([0, 1], [0, 1], 'k:', alpha=0.3)
axes[2].set_xlabel('False Positive Rate')
axes[2].set_ylabel('True Positive Rate')
axes[2].set_title('ROC Curves Comparison')
axes[2].legend(loc='lower right')

plt.tight_layout()
plt.savefig('./plots/lstm_before_after_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Feature Importance Analysis

```python
# feature importance from CatBoost (LSTM-enhanced)
importance = cat_lstm_model.get_feature_importance()
feat_imp = pd.DataFrame({
    'feature': feature_names_lstm,
    'importance': importance
}).sort_values('importance', ascending=False)

# mark LSTM features
feat_imp['type'] = feat_imp['feature'].apply(lambda x: 'LSTM' if x.startswith('lstm_emb') else 'Original')

print("Top 20 Features (LSTM-Enhanced Model):")
print("-" * 50)
for i, row in feat_imp.head(20).iterrows():
    marker = " [LSTM]" if row['type'] == 'LSTM' else ""
    print(f"  {row['feature']}: {row['importance']:.2f}{marker}")
```

```python
# compare feature importance: original vs LSTM
lstm_importance = feat_imp[feat_imp['type'] == 'LSTM']['importance'].sum()
orig_importance = feat_imp[feat_imp['type'] == 'Original']['importance'].sum()

print(f"\nFeature Type Contribution:")
print(f"  Original features: {orig_importance:.1f} ({orig_importance/(orig_importance+lstm_importance)*100:.1f}%)")
print(f"  LSTM embeddings:   {lstm_importance:.1f} ({lstm_importance/(orig_importance+lstm_importance)*100:.1f}%)")

# plot
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# top 15 features
top_15 = feat_imp.head(15)
colors = ['#F18F01' if t == 'LSTM' else '#2E86AB' for t in top_15['type']]
axes[0].barh(range(len(top_15)), top_15['importance'].values, color=colors)
axes[0].set_yticks(range(len(top_15)))
axes[0].set_yticklabels(top_15['feature'].values)
axes[0].invert_yaxis()
axes[0].set_xlabel('Importance')
axes[0].set_title('Top 15 Features (Blue=Original, Orange=LSTM)')

# feature type pie chart
axes[1].pie([orig_importance, lstm_importance], 
            labels=['Original Features', 'LSTM Embeddings'],
            autopct='%1.1f%%',
            colors=['#2E86AB', '#F18F01'])
axes[1].set_title('Feature Type Contribution')

plt.tight_layout()
plt.savefig('./plots/lstm_feature_importance.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5.1 SHAP Analysis for LSTM-Enhanced Models

```python
# SHAP Analysis for LSTM-Enhanced Models
import shap

# Sample data for SHAP (full data too slow)
sample_size = 2000
np.random.seed(42)
sample_idx = np.random.choice(len(X_test_lstm), sample_size, replace=False)
X_sample = X_test_lstm[sample_idx]

print(f"Computing SHAP values for {sample_size} samples...")
print("=" * 60)

def get_shap_values(model, model_name, X_sample, feature_names):
    """Get SHAP values with fallback to native importance."""
    try:
        explainer = shap.TreeExplainer(model)
        shap_values = explainer.shap_values(X_sample)
        
        if isinstance(shap_values, list):
            shap_values = shap_values[1]
        
        print(f"{model_name}: SHAP computed successfully")
        return shap_values, explainer
    except Exception as e:
        print(f"{model_name}: SHAP failed ({str(e)[:50]}...)")
        return None, None

# Compute SHAP for each LSTM-enhanced model
lgb_shap, lgb_explainer = get_shap_values(lgb_lstm_model, "LightGBM+LSTM", X_sample, feature_names_lstm)
cat_shap, cat_explainer = get_shap_values(cat_lstm_model, "CatBoost+LSTM", X_sample, feature_names_lstm)
xgb_shap, xgb_explainer = get_shap_values(xgb_lstm_model, "XGBoost+LSTM", X_sample, feature_names_lstm)
```

```python
# SHAP Summary Plots - Each model in separate figure for clarity
shap_available = []

if lgb_shap is not None:
    shap_available.append(('LightGBM+LSTM', lgb_shap))
if cat_shap is not None:
    shap_available.append(('CatBoost+LSTM', cat_shap))
if xgb_shap is not None:
    shap_available.append(('XGBoost+LSTM', xgb_shap))

if len(shap_available) > 0:
    print(f"\nGenerating SHAP Summary Plots for {len(shap_available)} LSTM-enhanced models...")
    print("=" * 60)
    
    # Create separate figures for each model (less congested)
    for name, shap_vals in shap_available:
        fig, ax = plt.subplots(figsize=(12, 10))
        plt.sca(ax)
        shap.summary_plot(shap_vals, X_sample, feature_names=feature_names_lstm, 
                         show=False, max_display=20)
        plt.title(f'{name} SHAP Summary - Top 20 Features', fontsize=14, pad=20)
        plt.tight_layout()
        
        # Save with model-specific name
        save_name = name.lower().replace('+', '_')
        plt.savefig(f'./plots/shap_summary_{save_name}.png', dpi=150, bbox_inches='tight')
        plt.show()
        print(f"Saved: ./plots/shap_summary_{save_name}.png")
    
    # Combined view with vertical stacking for better clarity
    n_models = len(shap_available)
    fig, axes = plt.subplots(n_models, 1, figsize=(14, 8*n_models))
    if n_models == 1:
        axes = [axes]
    
    for ax, (name, shap_vals) in zip(axes, shap_available):
        plt.sca(ax)
        shap.summary_plot(shap_vals, X_sample, feature_names=feature_names_lstm, 
                         show=False, max_display=15)
        ax.set_title(f'{name} SHAP Summary', fontsize=12)
    
    plt.tight_layout()
    plt.savefig('./plots/shap_summary_lstm_all_models.png', dpi=150, bbox_inches='tight')
    plt.show()
    print("\nSaved combined plot: ./plots/shap_summary_lstm_all_models.png")
else:
    print("SHAP summary plots not available - using native feature importance instead")
```
